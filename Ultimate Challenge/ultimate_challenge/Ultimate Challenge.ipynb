## Part 1

```python
logins = pd.read_json('logins.json')
logins.index = pd.to_datetime(logins['login_time'],format="%Y/%m/%d/%H/%m/%s")
```

```python
#logins['Sum'] = 1
logins_15min = logins.resample('15min').count()
```

```python
plt.rcParams["figure.figsize"] = (10,6)
logins_15min.plot()
plt.xlabel("Monthly 15 minute average")
plt.ylabel("Total number of logins")
```

```python
logins_1hour = logins.resample('1H').count()
```

```python
plt.rcParams["figure.figsize"] = (10,6)
logins_1hour.plot()
```

```python
logins =  logins.set_index('login_time')
logins['days'] = list(logins.index.dayofweek)
```

```python
logins.groupby(logins.index.hour).mean().plot(kind='bar', 
                                                                title='Average Logins per Hour Days',
                                                                color='#1d4991')
```

```python
logins_weekdays = logins[logins['days'].isin([0,1,2,3,4])]
logins_weekdays.groupby(logins_weekdays.index.hour).mean().plot(kind='bar', 
                                                                title='Average Logins per Hour Days',
                                                                color='#1d4991')
```

```python
def plot_one_week(start_date):
    plt.figure(figsize=(14,14))
    for day in range(1,8):
        plot_date = str(datetime.strptime(start_date, '%Y-%m-%d') + (timedelta(day - 1)))[:11]
        plt.subplot(3,3,day)
        plt.plot(logins_15min[plot_date]['login_time'])
        plt.xticks([plot_date + '01:00:00', plot_date + '12:00:00', plot_date + '23:00:00'], 
                   ['1AM', 'Noon', '11PM'])
        plt.title('Logins for ' + datetime.strftime(datetime.strptime(start_date, '%Y-%m-%d') + (timedelta(day - 1)), 
                                                    '%A %B, %d %Y'))
    plt.show()
```

```python
plot_one_week('1970-03-01')
```
